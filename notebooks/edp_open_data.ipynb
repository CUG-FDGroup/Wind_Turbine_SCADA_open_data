```python

data_path = f'../data/edp/' # path relative to notebook, adjust if different for you

dct_scada = load_scada_data(data_path)
dct_logs = load_scada_logs(data_path)
dct_annot = load_annotations(data_path)

dct_community = load_community_annotations(f'../communiy_annotations/edp_community_annotations.csv')
```

```python
df_data_overview = generate_data_overview(dct_scada, dct_logs)
```

```python
for trb_id in list(dct_scada.keys()):
    
    # function for overview with power curve, wind rose and data availability chart
    plot_overview_cockpit(data_path, dct_scada, df_data_overview, trb_id, freq_='10T')
    
    plot_summary(trb_id, df_data_overview.loc[trb_id], 
                 dct_annot[trb_id].drop('Turbine_ID',axis=1))
    
    if trb_id in dct_community.keys():
        plot_community_annotations(dct_community[trb_id], dct_scada[trb_id], dct_logs[trb_id])
    else:
        print('=' * 110 +' \n \033[1m \nCommunity Annotations: NONE SO FAR\n \033[1m \n')
```
