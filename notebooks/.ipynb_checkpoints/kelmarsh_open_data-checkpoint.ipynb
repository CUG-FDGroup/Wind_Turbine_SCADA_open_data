```python
from helper_functions import *
from community_plots import *

%matplotlib notebook
```

```python
dataset='edp'

data_path = f'../data/' # path relative to notebook, adjust if different for you

lst_scada_files = ['Wind-Turbine-SCADA-signals-2016.csv',  # adjust if named differently
                   'Wind-Turbine-SCADA-signals-2017_0.csv']

lst_log_files = ['Wind-Turbines-Logs-2016.csv',  # adjust if named differently
                 'Wind Turbines Logs 2017.csv']

lst_files_annot = ['Historical-Failure-Logbook-2016.csv', # adjust if named differently
                   'opendata-wind-failures-2017.csv']

dct_scada = load_scada_data(data_path, lst_scada_files, dataset)
dct_logs = load_scada_logs(data_path, lst_log_files, dataset)
dct_annot = load_annotations(data_path, lst_files_annot, dataset)

dct_community = load_community(f'../communiy_annotations/{dataset}_community_annotations.csv')
```

```python
df_data_overview = generate_data_overview(dct_scada, dct_logs)
```

```python
for trb_id in list(dct_scada.keys()):
    
    # function for overview with power curve, wind rose and data availability chart
    plot_overview_cockpit(data_path+f'{dataset}/', dataset, dct_scada, df_data_overview, trb_id, freq_='10T')
    
    plot_summary(trb_id, df_data_overview.loc[trb_id], 
                 dct_annot[trb_id].drop('Turbine_ID',axis=1))
    
    if trb_id in dct_community:
        plot_community_annotations(dataset, dct_community[trb_id], dct_scada[trb_id], dct_logs[trb_id])
```
